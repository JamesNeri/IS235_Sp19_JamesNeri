```python
#USE PANDAS TO WORK WITH CSV FILES
import pandas as pd
#OPEN THE CSV FILE "sentiment_analysis.csv"
#FIRST COLUMN IS A WORD, SECOND COLUMN IS A RATING, THIRD COLUMN IS THE PRICE
file=pd.read_csv("sentiment_analysis.csv")
Buy_count = 0
Sell_count = 0
DoNothing_count = 0
#IF THE WORD HAS AN "E" IN IT, USE A FACTOR OF 1000
#IF THE WORD DOES NOT HAVE AN "E" IN IT, USE A FACTOR OF 2000
#RATING * FACTOR = TARGET PRICE
length = file.index.size
for i in range(length):
    word = (file.at[i,"Sentiment"])
    rating = float(file.at[i,"Rating"])
    price = float(file.at[i,"Price"])
    
    if "e" in word and "w" in word:
        Factor = 3000
        target_price = (Factor)*(rating)
    elif "t" in word and "r" in word:
        Factor = 4000
        target_price = (Factor)*(rating)
    elif "q" not in word and "z" not in word:
        Factor=5000
        target_price = (Factor)*(rating)
    elif "e" in word:
        Factor = 1000
        target_price = (Factor)*(rating)        
    else:
        Factor = 2000
        target_price = (Factor)*(rating)
    
    #IF TARGET PRICE IS HIGHER, BUY
    #IF TARGET PRICE IS LOWER, SELL (SHORT)
    #IF TARGET PRICE IS EQUAL, DO NOTHING

    if target_price > price:
        print ("Buy", word, Factor)
        Buy_count = Buy_count+1
    elif target_price < price:
        print ("Sell (short)", word, Factor)
        Sell_count = Sell_count+1
    elif target_price==price:
        print ("Do Nothing",word, Factor)
        DoNothing_count = DoNothing_count+1
#FOR EACH ROW IN THE CSV FILE
    #PRINT THE WORD
    #PRINT WETHER IT HAS AN E OR NOT
    #PRINT THE FACTOR USED FOR THE ROW
    #PRINT THE CURRENT PRICE
    #PRINT THE TARGET PRICE
    #PRINT THE DECISION
#PRINT NUMBER OF BUYS, SELLS, NOTHINGS
print ("Buy_count: ", Buy_count, "Sell_count: ", Sell_count, "DoNothing_count : ", DoNothing_count)
```
